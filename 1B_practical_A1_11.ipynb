```python
import pandas as pd
```

```python
data1 = {
    "ID" : [101, 102, 103],
    "Name" : ["Alice Smith", "Bob Jones", "Charlie Brown"],
    "Salary" : [50000, 60000, 55000]
}

df1 = pd.DataFrame(data1)

display(df1)
```

```python
data2 = {
    "Emp_ID" : [201, 202, 103],
    "FullName" : ["Alice S", "Robert Jones", "Charlie Brown"],
    "Income" : [50000, 60000, 50000]
}

df2 = pd.DataFrame(data2)
display(df2)
```

```python
df2_renamed = df2.rename(columns = {"Emp_ID" : "ID",
                                    "FullName" : "Name",
                                    "Income" : "Salary"
                                    })

display(df2_renamed)
```

```python
from fuzzywuzzy import process

def match_names(name, choices, threshold = 75):
  match,score = process.extractOne(name, choices)
  if score >= threshold:
    return match
  else:
    return None


```

```python
df2_renamed["Matched_Name"] = df2_renamed["Name"].apply(lambda x:match_names(
    x, df1["Name"].tolist()
))
```

```python
display(df2_renamed)
```

```python
merged_df = pd.merge(df1, df2_renamed, left_on = 'Name', right_on = 'Matched_Name', how = 'outer', suffixes = ('_src1', '_src2'))
```

```python
display(merged_df)
```

```python
merged_df["final_salary"] = merged_df.apply( lambda row: row['Salary_src1'] if pd.notnull(row['Salary_src2']) else False, axis = 1)
display(merged_df)
```

```python
merged_df["Salary_conflict"] = merged_df.apply(
    lambda row: row['Salary_src1'] != row['Salary_src2']
    if pd.notnull(row['Salary_src1']) and pd.notnull(row['Salary_src2'])
    else False,
    axis = 1
    )
display(merged_df)
```

```python
final_df = merged_df[['ID_src1', 'Name_src1', 'final_salary', 'Salary_conflict']]
final_df.columns = ['ID', 'Name', 'Salary', 'Salary_conflict']

display(final_df)
```
